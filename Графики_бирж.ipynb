## Загрузка данных с биржи

```python
data = yf.download('AAPL', '2020-01-01','2022-12-05') #Период

import matplotlib.pyplot as plt

data['Open'].plot()
data['High'].plot()
data['Low'].plot()
data['Close'].plot()
```

# Свечной график

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=data.index,
                                     open=data['Open'],
                                     high=data['High'],
                                     low=data['Low'],
                                     close=data['Close'])])

fig.show(0)
```

```python
from numpy.core.fromnumeric import cumprod
import pandas as pd

tickers_list= ['AAPL', 'UBI.PA','INTC']
data= pd.DataFrame(columns=tickers_list)

for ticker in tickers_list:
  data[ticker] = yf.download(ticker, '2020-01-01','2022-12-05')['Adj Close']

data.head()

((data.pct_change()+1).cumprod()).plot(figsize=(10,7))
```
